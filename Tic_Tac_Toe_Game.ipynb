```python
import copy
import pygame
from pygame.locals import *

currentplayer  = 1  
grid =[[None for j in range(3)] for i in range(3)]

'''
This is the class to realise the IA part of the game
'''
class Game():
    
    def __init__ (self):
        self.gridgame = [[0 for j in range(3)] for i in range(3)]
        self.gridofposs = [0, 1, 2, 3, 4, 5, 6, 7, 8]
        
    '''
    Check if where the player wants to play is free, if it's true it put his pawn 
    '''
    def play(self, player,index ):
                         
        if (index>=0) and (index<3 ):
            x=0
        elif (index>=3) and (index<6 ):
            x=1
        elif (index>=6) and (index<9 ):
            x=2     
        if (index==0) or (index==3) or (index==6):
            y=0
        elif (index==1) or (index==4) or (index==7):
            y=1
        elif (index==2) or (index==5) or (index==8):
            y=2       
        
        if index>8:
            return ()
        if player != 2 and player != 1:
            return ()
        if self. gridgame[x] [y] == 0:
            self.  gridgame[x] [y] = player
   
    '''
    Thoses functions are the diffrents conditionals function for the diffrents levelm of the minmax 
    level+thelevel+1or 2 if it's the user case of not in the minmax algorithm
    '''
    def level31(self,value,pos,v):
        if value[pos]>v:
            return True
        else:
            return False
    def level32(self,value,pos,v):
        if value[pos]<v:
            return True
        else:
            return False
        
    def level21(self,value,pos,v):
        if value[pos]==v:
            return True
        else:
            return False
    def level22(self,value,pos,v):
        if value[pos]==v:
            return True
        else:
            return False
    def level11(self,value,pos,v):
        if value[pos]<v:
            return True
        else:
            return False
    def level12(self,value,pos,v):
        if value[pos]>v:
            return True
        else:
            return False        
        
    '''
    The minmax Algorithm
    '''    
    def minmax(self,  user, gridofposs, root,level):
        
        
        if  user==False:
            player=1
            player2=2
        else:
            player=2
            player2=1
        
        if self.win(player):
            if user:
                return (-1)
            else:
                return (+1)
            
        if gridofposs== []:
            return 0
        value = [0 for i in range (9)]
        for move in gridofposs:
            newPos = copy.deepcopy (self)
            newPos. play (player2, move)
            cp = copy.deepcopy (gridofposs)
            cp. remove (move)
            value [move] = newPos. minmax( not user, cp, False,level)
            v = value[gridofposs [0]]
            themove =gridofposs [0]
        
        for move in gridofposs:
            if level ==3:
                if user:
                     if self.level31(value,move,v):
                        v = value [move]
                        themove = move
                else:
                     if self.level32(value,move,v):
                        v = value [move]
                        themove= move
            elif level==2:                
                if user:
                    if self.level21(value,move,v):
                        v = value [move]
                        themove = move
                else:
                    if self.level22(value,move,v):
                        v = value [move]
                        themove= move
            elif level==1:              
                if user:
                    if self.level11(value,move,v):
                        v = value [move]
                        themove= move
                else:
                     if self.level12(value,move,v):
                        v = value [move]
                        themove = move
        if root:
            return (themove)
        else:
            return (v)
        
    '''
    Check if the player is the winner
    '''
    def win(self, player):
        for i in range (0, 3):
            if (self. gridgame[i] == [player,player, player]):
                return (True)
        for i in range (0, 3):
            if (self. gridgame [0] [i] == player) and (self. gridgame [1] [i] ==player) and (self. gridgame [2] [i] == player):
                return (True)
        if (self. gridgame [0] [0] == player) and (self. gridgame[1] [1] == player) and (self. gridgame [2] [2] == player):
            return (True)
        if (self. gridgame[0] [2] == player) and (self. gridgame [1] [1] == player) and (self. gridgame[2] [0] == player):
            return (True)
        return (False)
    
'''
THE PYGAME INTEFACE PART
''' 
    
'''
Update the grid when a new mode of game is selected, it deletes the part a draw the grid and draw a new one
'''
def update(window):
    window.fill((128, 128, 128), (0, 0, 350,350 ))
    ##vertical
    pygame.draw.line (window, (0,0,0), (50, 50), (50, 355), 10)
    pygame.draw.line (window, (0,0,0), (150, 50), (150, 350), 10)
    pygame.draw.line (window, (0,0,0), (250, 50), (250, 350), 10)
    pygame.draw.line (window, (0,0,0), (350, 50), (350, 350), 10)
    ##Horizontal
    pygame.draw.line (window, (0,0,0), (46, 50), (355,50), 10)
    pygame.draw.line (window, (0,0,0), (50, 150), (350, 150), 10)
    pygame.draw.line (window, (0,0,0), (50, 250), (350, 250), 10)
    pygame.draw.line (window, (0,0,0), (50, 350), (355, 350), 10)
    
'''
Initialise all of the pygame inteface, buttonns ,texts, and the picture
'''
def initialisation(window):
    window.fill((128, 128, 128))

    pygame.font.init()
   
    myfont = pygame.font.SysFont('arial', 24)
    txt = myfont.render("1 VS 1", 1, (100 ,30, 10))
    window.fill ((192, 192, 192), (380, 50, 75,30 ))
    window.blit(txt, (390, 50))
    
    myfont = pygame.font.SysFont('arial', 24)
    txt = myfont.render("VS IA Level 1", 1, (100 ,30, 10))
    window.fill ((0, 128, 128), (380, 90, 150,30 ))
    window.blit(txt, (390, 90))
    
    myfont = pygame.font.SysFont('arial', 24)
    txt = myfont.render("VS IA Level 2", 1, (100 ,30, 10))
    window.fill ((0, 128, 128), (380, 130, 150,30 ))
    window.blit(txt, (390, 130))
    
    myfont = pygame.font.SysFont('arial', 24)
    txt = myfont.render("VS IA Level 3", 1, (100 ,30, 10))
    window.fill ((0, 128, 128), (380, 170, 150,30 ))
    window.blit(txt, (390, 170))
    
    myfont = pygame.font.SysFont('arial', 22)
    txt = myfont.render("Welcome on Maxime & Joakim 's Tic Tac Toe", 1, (0, 51, 153))
    window.blit(txt, (535, 130))
    myfont = pygame.font.SysFont('arial', 20)
    txt = myfont.render("-Click on the differents modes to play", 1, (0, 51, 153))
    window.blit(txt, (540, 160))
    myfont = pygame.font.SysFont('arial', 20)
    txt = myfont.render("-At any time, you can click on a mode to start", 1, (0, 51, 153))
    window.blit(txt, (540, 190))
    myfont = pygame.font.SysFont('arial', 20)
    txt = myfont.render("a new game and select the mode", 1, (0, 51, 153))
    window.blit(txt, (540, 210))
    
   
    logo=pygame.image.load("logoesilv.jpeg").convert()
    logo=pygame.transform.scale(logo,(1280/4,384/4)) 
    window.blit(logo, (560, 15))

    return window
'''
Change the game satut by the new enter in parameters
'''
def gamestatue(board,message):
  
    mess=str(message)
    pygame.font.init()
    myfont = pygame.font.SysFont('arial', 30)
    txt = myfont.render(mess, 1, (51, 0, 17))

    board.fill ((128, 128,128), (160, 10, 350,35 ))
    board.blit(txt, (160, 10))
    
'''
Return the row and the column of the grid game if the player click inside, else it return -1
''' 
def boardPosition (mouseX, mouseY):  
    
    if (mouseY < 150):
        row = 0
    elif (mouseY < 250):
        row = 1
    elif (mouseY < 350):
        row = 2
    else:
        row=-1
 
    if (mouseX < 150):
        col = 0
    elif (mouseX < 250):
        col = 1
    elif (mouseX < 350):
        col = 2
    else:
        col=-1
    
    return (row, col)

'''
Draw the pawn on the grid when a player put one, and add this on the general grid to control the game
'''
def drawpawn (board,Row, Col, TypeOfPawn):
    
    indexXcenter = (((Col)+1) * 100) 
    indexYcenter = (((Row)+1) * 100) 

  
    if (TypeOfPawn == 2):
        pygame.draw.circle (board, (0,0,255), (indexXcenter, indexYcenter), 40, 5)
    else:
        pygame.draw.line (board, (255,0,0), (indexXcenter - 30, indexYcenter - 30),  (indexXcenter+ 30, indexYcenter + 30), 7)
        pygame.draw.line (board, (255,0,0), (indexXcenter+ 30, indexYcenter -30),  (indexXcenter - 30, indexYcenter+ 30), 5)

    grid [Row][Col] = TypeOfPawn

'''
Initialisation of the click on a game, the player have to click on a button to launch a game, he can't add pawn on the grid
'''  
def initclickevent():
    (mouseX, mouseY) = pygame.mouse.get_pos()
    global grid
    if 380<mouseX<455 and 50<mouseY<80:
       
        
        grid=[[None for j in range(3)] for i in range(3)]
        currentplayer=1
        newgame()
    if 380<mouseX<530 and 90<mouseY<120:
        grid=[[None for j in range(3)] for i in range(3)]
        newgame2(1) 
    if 380<mouseX<530 and 130<mouseY<160:
        grid=[[None for j in range(3)] for i in range(3)]
        newgame2(2)
    if 380<mouseX<530 and 170<mouseY<200:
        grid=[[None for j in range(3)] for i in range(3)]
        newgame2(3)
    
'''
Click event manage the click on 1vs1 game
'''
def clickevent(board): 
    
    global grid, currentplayer
    
    (mouseX, mouseY) = pygame.mouse.get_pos()
    
    (row, col) = boardPosition (mouseX, mouseY)
    if 380<mouseX<455 and 50<mouseY<80:
       
        
        grid=[[None for j in range(3)] for i in range(3)]
        currentplayer=1
        newgame()
    if 380<mouseX<530 and 90<mouseY<120:
        grid=[[None for j in range(3)] for i in range(3)]
        newgame2(1)
    if 380<mouseX<530 and 130<mouseY<160:
        grid=[[None for j in range(3)] for i in range(3)]
        newgame2(2)
    if 380<mouseX<530 and 170<mouseY<200:
        grid=[[None for j in range(3)] for i in range(3)]
        newgame2(3)
    if (row!=-1 and col!=-1):
        
   
        if ((grid[row][col] == 1) or (grid[row][col] == 2)):
        
            return    
        drawpawn (board, row, col, currentplayer)  
        if (currentplayer == 1):
            gamestatue(gamevisu,'It is the O turn')
            currentplayer = 2
        else:
            gamestatue(gamevisu,'It is the X turn')
            currentplayer = 1

'''
Click event manage the click on VS IA game, for level 1,2 and 3 enter in parameters
'''          
def clickevent2(board,jeu,gridofposs,level):
   
    (mouseX, mouseY) = pygame.mouse.get_pos()
    global grid
    if 380<mouseX<455 and 50<mouseY<80:        
        grid=[[None for j in range(3)] for i in range(3)]
        currentplayer=1
        newgame()
    if 380<mouseX<530 and 90<mouseY<120:
        grid=[[None for j in range(3)] for i in range(3)]
        newgame2(1)
    if 380<mouseX<530 and 130<mouseY<160:
        grid=[[None for j in range(3)] for i in range(3)]
        newgame2(2)
    if 380<mouseX<530 and 170<mouseY<200:
        grid=[[None for j in range(3)] for i in range(3)]
        newgame2(3)
    
    
    (row, col) = boardPosition (mouseX, mouseY)
    if (row==0) and (col==0):
            
            case=0
    elif (row==0) and (col==1):
        case=1
    elif (row==0) and (col==2):
        case=2
            
    elif (row==1) and (col==0):
        case=3
    elif (row==1) and (col==1):
        case=4
    elif (row==1) and (col==2):
        case=5
    elif (row==2) and (col==0):
        case=6
    elif (row==2) and (col==1):
        case=7
    elif (row==2) and (col==2):
        case=8
    else:
        case=-1
    if case in gridofposs:
                       
        jeu. play (1, case )
        drawpawn(board, row, col, 1)
       
        if  gridofposs!=[]:
            gridofposs.remove(case)
      
    
        if  gridofposs==[]:
            return
        gameIsWin(board)
        isfull(board)
        
        if gameIsWin(board):
            gamestatue(gamevisu,'You Win')
            return
        caseAjouer = jeu. minmax (False, gridofposs, True,level)
    
   
        gridofposs. remove (caseAjouer)
        jeu. play (2, caseAjouer)
        
        drawpawn (board, caseAjouer//3, caseAjouer%3, 2)
        if gameIsWin(board):
            gamestatue(gamevisu,'You Lose')
            return
        gamestatue(gamevisu,'It is your turn : You can play')

            
'''
Check if the game is win and draw the line in consequence if it's true
'''    
def gameIsWin(board):
   
    global grid
                
    for row in range (0, 3):
        if ((grid [row][0] == grid[row][1] == grid[row][2]) and  (grid [row][0] is not None)):
            
            if  grid[row][0]==1:
                color=(250,0,0)
            else:
                color=(0,0,250)
                
            pygame.draw.line (board, color, (50, (row + 1)*100 ),(350, (row + 1)*100 ), 20)
            
            return True
 
    for col in range (0, 3):
        if (grid[0][col] == grid[1][col] == grid[2][col]) and (grid[0][col] is not None):
          
            if  grid[0][col]==1:
                color=(250,0,0)
            else:
                color=(0,0,250)
            pygame.draw.line (board, color, ((col + 1)* 100, 75),  ((col + 1)* 100, 325), 20)
            return True
            

    
    if (grid[0][0] == grid[1][1] == grid[2][2]) and (grid[0][0] is not None):
    
        if  grid[0][0]==1:
                color=(250,0,0)
        else:
                color=(0,0,250)
        
        pygame.draw.line (board, color, (75,75), (325, 325), 20)
        return True
    if (grid[0][2] == grid[1][1] == grid[2][0]) and (grid[0][2] is not None):
        
        if  grid[0][2]==1:
                color=(250,0,0)
        else:
                color=(0,0,250)
        pygame.draw.line (board, color, (325, 75), (75, 325), 20)
        return True
    return False


'''
Check if the grid is full
'''
def isfull(board): 
    global grid
    if gameIsWin(board)!=True:
        cmpt=0
        for i in range (0, 3):
            for j in range (0, 3):
                if grid[i][j]!=None:
                    cmpt+=1
        if cmpt==9:
            gamestatue(gamevisu,"It's a Draw !!!")
            return True
        else:
            return False
    else:
        return False
    

    
'''
MANAGEMENT OF THE INTERFACE
'''    
    
'''
Initialisation of the interface
'''
pygame.init()    
pygame.display.set_caption (' ADSA Final Workshop VO & TAZI Tic-Tac-Toe')
pygame.init()
window = pygame.display.set_mode ((900,400))
window = window.convert()
gamevisu = initialisation (window)

       
            

'''
Newgame matches with the type of game 1Vs1 and will launch the clickevent
'''         
def newgame():
    
    run = True
    update(gamevisu)
    gamestatue(gamevisu,'1 VS 1: X player first')
    while (run == True):
       
        for event in pygame.event.get():
            if event.type is QUIT:
                run = False
                pygame.quit()
                break

            elif event.type is MOUSEBUTTONDOWN:
         
                clickevent(gamevisu)

       
            gameIsWin(gamevisu)
            isfull(gamevisu)
            if gameIsWin(gamevisu):
                if currentplayer==1:
                    
                    gamestatue(gamevisu,'Player O Win')
                else: 
                    
                    gamestatue(gamevisu,'Player X Win')
                return
        
            bd = pygame.display.set_mode ((900,400))
            bd.blit (gamevisu, (0, 0))
            pygame.display.flip()
            
'''
Newgame2 matches with the type of game VS AI and will launch the clickevent2
'''            
def newgame2(level):
    jeu = Game()
    gridofposs = [0, 1, 2, 3, 4, 5, 6, 7, 8]
    
   
    run = True
    update(gamevisu)
    gamestatue(gamevisu,'VS IA Level '+str(level)+' -> You play first')
    while (run == True):
       
        for event in pygame.event.get():
            if event.type is QUIT:
                run = False
                pygame.quit()
                break

             
            elif event.type is MOUSEBUTTONDOWN:
         
                clickevent2(gamevisu,jeu,gridofposs,level)

      
            gameIsWin(gamevisu)
            isfull(gamevisu)   
            if gameIsWin(gamevisu):
                return
       
           
            bd = pygame.display.set_mode ((900,400))
            bd.blit (gamevisu, (0, 0))
            pygame.display.flip()
            
'''
Newwindow is the begining of the interface, in initiate the pygame intaface with initclickevent
'''          
def newwindow():
    run = True
    update(gamevisu)
    gamestatue(gamevisu,'Select a Game mode')
    while (run == True):
       
        for event in pygame.event.get():
            if event.type is QUIT:
                run = False
                pygame.quit()
                break
                
            elif event.type is MOUSEBUTTONDOWN:
                initclickevent()
                
            bd = pygame.display.set_mode ((900,400))
            bd.blit (gamevisu, (0, 0))
            pygame.display.flip()       
'''
Launch the inteface 
'''            
newwindow()  
```
